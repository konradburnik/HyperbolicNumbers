```python
import numpy
import matplotlib.cm as cm
import matplotlib.pylab as plt
from hyperbolic import *

class HyperbolicFractal:
    def __init__(self):
        self.hits = 0
        self.cache = {}
        
    def f(self, z, c):
        if (z, c) in self.cache:
            self.hits += 1
            return self.cache[(z, c)]
        w = z
        total = 0
        while mag(w) < 2.0 and total < 300:
            w = w**2 + c
            total += 1
        self.cache[(z, c)] = total
        return self.cache[(z, c)]

    # Draw a hyperbolic fractal
    def show(self, scale, centerx, centery, rangex, rangey, c):
        self.hits = 0
        totals = (centerx+rangex - (centerx-rangex)) * (centery+rangey - (centery-rangey))
        # TODO: Add caching of already calculated points
        t = [[self.f((1.0/(2**scale)) * HyperbolicNumber(x,y), c) for x in range(centerx-rangex, centerx+rangex)] for y in range(centery-rangey, centery+rangey)]
        matrix = numpy.matrix(t)
        fig = plt.figure()
        plt.imshow(matrix, interpolation='bilinear', cmap=cm.RdYlGn, origin='center')
        print ("Total: {} Cache hits: {} Cache hit ratio: {}".format(totals, self.hits, (1.0 * self.hits)/totals))
        plt.show()
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
H = HyperbolicFractal()
```

```python
for x in range(-10, 10):
    for y in range(-10, 10):
        print("x:", x, "y:", y)
        H.show(3, -80, 10, 100, 100, HyperbolicNumber((x+y)/2**3, (x-y)/2**3))
```
